## Παρατηρησεις
Απο τον πινακα με τα missing values βλεπουμε τα εξης:
- Παρατηρουμε οτι με εξαιρεση την κολωνα με τις ημερομηνιες σε ολες τις αλλες κολωνες λειπουν πανω απο 9Κ παρατηρησεις. Αυτο αντιστοιχει σε περιπου 2% των παρατηρησεων επι του συνολου σε καθε κολωνα 
- Να τονιστει εδω οτι οι ~9Κ παρατηρησεις που λειπουν απο καθε κολωνα οφειλονται στον μηνα Μαιο και αντιστοιχουν περιπου στο 14% των παρατηρησεων του Μαιου. Οι υπολοιποι μηνες εχουν πρακτικα μηδαμινα ΝΑ σε σχεση με το συνολο των παρατηρησεων. Πιο συγκεκριμενα τα ΝΑ αυτα του Μαιου, αντιστοιχουσαν σε καποιο ERROR (string) το οποιο μετα την μετατροπη των παρατηρησεων σε αριθμητικες μετατραπηκαν σε ΝΑ.
- Σε ολες τις κολωνες λειπει πανω κατω ο ιδιος αριθμος παρατηρησεων - ομοιομορφη κατανομη των ΝΑ μεταξυ των μεταβλητων. Αυτο σημαινει οτι πιθανον καποια συγκεκριμενα instancies να μην εχουν values σε ολες τις κολωνες (οποτε η αφαιρεση παρατηρησεων θα ειναι η μικροτερη δυνατη). Οπως και να χει δεν μας νοιαζει διοτι το ποσοστο τους ειναι πολυ χαμηλο (~2%). 
- Ο Οκτωβριος και ο Ιουλιος δεν εχουν καθολου, ο Σεπτεμβριος εχει 1 missing value σε καποιες κολωνες, ο Αυγουστος μεταξυ 0 και 6, ενω ο Ιουνιος γενικα εχει πληροτητα πανω απο 99,9%.
 
***
**Steps:**
1. Αρχικα θα κανουμε αφαιρεση των ΝΑ με threshold = 2. Αυτο σημαινει οτι για να επιβιωσει μια γραμμη πρεπει να εχει τουλαχιστον 2 non-NA values. Η μια ειναι σιγουρα η ημερομηνια (και μια πιθανη δευτερη ο μυλος). 
3. Τσεκαρουμε παλι τα ΝΑ.
```python
def crosscorr(datax, datay, lag=0, wrap=False):
    """ Lag-N cross correlation. 
    Shifted data filled with NaNs 
    
    Parameters
    ----------
    lag : int, default 0
    datax, datay : pandas.Series objects of equal length
    Returns
    ----------
    crosscorr : float
    """
    if wrap:
        shiftedy = datay.shift(lag, fill_value=0)
        shiftedy.iloc[:lag] = datay.iloc[-lag:].values
        shiftedy = shiftedy[~np.isnan(shiftedy)]
        return mean_squared_error(datax, shiftedy)
    else: 
        return mean_squared_error(datax,datay.shift(lag, fill_value=0))
```

```python
def crosscorr(datax, datay, lag=0, wrap=False):
    """ Lag-N cross correlation. 
    Shifted data filled with NaNs 
    
    Parameters
    ----------
    lag : int, default 0
    datax, datay : pandas.Series objects of equal length
    Returns
    ----------
    crosscorr : float
    """
    if wrap:
        shiftedy = datay.shift(lag, fill_value=0)
        shiftedy.iloc[:lag] = datay.iloc[-lag:].values
        shiftedy = shiftedy[~np.isnan(shiftedy)]
        return mean_squared_error(datax, shiftedy)
    else: 
        return mean_squared_error(datax,datay.shift(lag, fill_value=0))
```
